```python
import pandas as pd
from datetime import datetime
import numpy as np

df_data = pd.read_csv('df_final_data.csv')
```

```python
df_data = df_data.sort_values(by=['client_id','visitor_id','visit_id','date_time']).reset_index(drop=True)
```

```python
def time_spend(row, next_row):
    formato = "%Y-%m-%d %H:%M:%S"
    if row['visitor_id'] == next_row['visitor_id'] and row['client_id'] == next_row['client_id'] and row['visit_id'] == next_row['visit_id']:
        if row['process_step'] == 'start' and next_row['process_step'] == 'step_1' :
            return abs(datetime.strptime(next_row['date_time'], formato) - datetime.strptime(row['date_time'], formato)).total_seconds()
        elif row['process_step'] == 'step_1' and next_row['process_step'] == 'step_2':
            return abs(datetime.strptime(next_row['date_time'], formato) - datetime.strptime(row['date_time'], formato)).total_seconds()
        elif row['process_step'] == 'step_2' and next_row['process_step'] == 'step_3':
            return abs(datetime.strptime(next_row['date_time'], formato) - datetime.strptime(row['date_time'], formato)).total_seconds()
        elif row['process_step'] == 'step_3' and next_row['process_step'] == 'confirm':
            return abs(datetime.strptime(next_row['date_time'], formato) - datetime.strptime(row['date_time'], formato)).total_seconds()
        else:
            return np.nan
    else:
        return np.nan

df_data['time_spend'] = np.nan
for i in range(len(df_data) - 1):
    df_data.loc[i, 'time_spend'] = time_spend(df_data.loc[i], df_data.loc[i + 1])
df_data
```

```python
def get_error(row, previous_row):
    if row['visitor_id'] == previous_row['visitor_id'] and row['client_id'] == previous_row['client_id'] and row['visit_id'] == previous_row['visit_id']:
        if row['process_step'] == 'start' and previous_row['process_step'] == 'step_1':
            return 1
        elif row['process_step'] == 'step_1' and previous_row['process_step'] == 'step_2':
            return 1
        elif row['process_step'] == 'step_2' and previous_row['process_step'] == 'step_3':
            return 1
        elif row['process_step'] == 'step_3' and previous_row['process_step'] == 'confirm':
            return 1
        else:
            return 0
    else:
            return 0

df_data['user_error'] = np.nan
for i in range(1,len(df_data)):
    df_data.loc[i-1, 'user_error'] = get_error(df_data.loc[i], df_data.loc[i - 1])
df_data
```

```python
df_data[['client_id','visitor_id','visit_id','date_time','user_error','process_step']].head(50)
```

```python
total = df_data.groupby('process_step')['client_id'].count().reset_index(name='count')

error_df = df_data.groupby('process_step')['user_error'].value_counts().reset_index()
error_aux = error_df[error_df['user_error'] == 1]
error = pd.merge(error_aux, total, on='process_step')
error.rename(columns={'process_step': 'process_step','user_error':'user_error','count_x':'total_errors','count_y':'total_steps'}, inplace=True)
error['error_rate'] = round(error['total_errors']/error['total_steps']*100, 2)
error.sort_values(by='error_rate', ascending=False)
```

```python
def age_to_category(age):
    category = ''
    if age < 10:
        category = 'young'
    elif 10 <= age < 20:
        category = 'teenager'
    elif 20 <= age < 50:
        category = 'adult'
    elif age >= 50:
        category = 'senior'
    return category

df_data['age_category'] = df_data['clnt_age'].apply(lambda x : age_to_category(x))
```

```python
def bal_to_category(bal):
    if bal < 10000:
        return 'Low'
    elif 10000 <= bal < 50000:
        return 'Medium'
    elif 50000 <= bal < 100000:
        return 'High'
    else:  # bal >= 100000
        return 'Very High'

df_data['balance_category'] = df_data['bal'].apply(lambda x : bal_to_category(x))
```
