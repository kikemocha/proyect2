```python
import pandas as pd
import numpy as np
import scipy.stats as st

df = pd.read_csv('data_final_1.csv')
df
```

```python
hypo1 = df.groupby(['Variation_x','process_step'])['client_id'].count().reset_index()
```

```python
total_control = hypo1[hypo1['Variation_x'] == 'Control'].groupby('Variation_x')['client_id'].sum()
```

```python
hypo1_Control = hypo1[hypo1['Variation_x'] == 'Control']
hypo1_Control['rate'] = hypo1_Control['client_id'].apply(lambda x : round((x / total_control)*100,2))
hypo1_Control
```

```python
total_test = hypo1[hypo1['Variation_x'] == 'Test'].groupby('Variation_x')['client_id'].sum()
```

```python
hypo1_Test = hypo1[hypo1['Variation_x'] == 'Test']
hypo1_Test['rate'] = hypo1_Test['client_id'].apply(lambda x : round((x / total_test)*100,2))
hypo1_Test
```

```python
pd.concat([hypo1_Test, hypo1_Control], axis=0)
```

```python
test_completion_rate = hypo1_Test[hypo1_Test['process_step'] == 'confirm']['rate']
control_completion_rate = hypo1_Control[hypo1_Control['process_step'] == 'confirm']['rate']
```

```python
# H0 --> TEST completion rate == CONTROL completion rate
# H1 --> TEST completion rate != CONTROL completion rate
test_completion_rate , control_completion_rate
# Podemos negar la hipotesis 0
```

```python
# 2
#H0 --> Diference between TEST completion rate and CONTROL completion rate is at least 5%
#H1 --> Diference between TEST completion rate and CONTROL completion rate is more than 5%
abs(float(test_completion_rate) - float(control_completion_rate))
# No Podemos negar la hipotesis 0
```

```python
# 3
# H0 --> Gender is important between TEST and CONTROL
# H1 --> Gender is NOT important between TEST and CONTROL
contingency_table = pd.crosstab(df['gendr'], df['Variation_x'])
chi2, p, dof, expected = st.chi2_contingency(contingency_table)

alpha = 0.05

if p > alpha:
    print(f"No podemos rechazar la hipotesis 0 el {100 - (alpha)*100}% de la veces")
else:
    print(f"Podemos rechazar la hipotesis 0 el {100 - (alpha)*100}% de la veces")
```

```python
# 4
# H0 --> El dinero que tenga el cliente es importante between TEST and CONTROL
# H1 --> El dinero que tenga el cliente NO es importante  important between TEST and CONTROL

contingency_table = pd.crosstab(df['balance_category'], df['Variation_x'])
chi2, p, dof, expected = st.chi2_contingency(contingency_table)

alpha = 0.05

if p > alpha:
    print(f"No podemos rechazar la hipotesis 0 el {100 - (alpha)*100}% de la veces")
else:
    print(f"Podemos rechazar la hipotesis 0 el {100 - (alpha)*100}% de la veces")
```
